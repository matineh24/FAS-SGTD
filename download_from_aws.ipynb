Get all data in oulu-npu dataset

```python
import tarfile
from os.path import join

tar = tarfile.open(join('OULU-NPU/', 'Dev_files.tar'), "r:")
tar.extractall()
tar.close()
```

```python
tar = tarfile.open(join('OULU-NPU/', 'Protocols.tar'), "r:")
tar.extractall()
tar.close()
```

Get the name of the files in oulu-npu dataset related to protocol1 dev-set

```python
lines = []
with open('Protocols/Protocol_1/Dev.txt') as f:
    lines = f.readlines()
files_i_want = []
count = 0
for line in lines:
    x = line.split(',')
    files_i_want.append(str(x[1][:8])+'.avi')
    files_i_want.append(str(x[1][:8])+'.txt')
```

Remove all files in dev set of oulu-npu dataset which are not used in protocol 1

```python
from os import listdir

files = [f for f in listdir('Dev_files/')]
for i in files:
  if not i in files_i_want:
    string = 'Dev_files/'+i
    !rm $string
```

Zip and save the new dev_set 

```python
!zip -r dev_files.zip Dev_files/
```
